```python
move_files = []
for name in xml_files:
    move_files.append(name.strip('.xml'))
```

```python
file_types = ['.xml', '.png']
n = 0
for file in move_files:
    for ext in file_types:
        if n < 15:
            shutil.move(file + ext, 'images/Test/')
        else:
            shutil.move(file + ext, 'images/Train/')
    n += 1
```
